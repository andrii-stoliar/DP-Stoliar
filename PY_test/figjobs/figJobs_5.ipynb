MRAC simulation raw experiment

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from figFcns_TeX import *

# ===== Params =====
figSaveDir = '../fig'
os.makedirs(figSaveDir, exist_ok=True)

figName = 'ts_ml_mrac_simulation_raw'

# ===== Load CSV =====
data_main = pd.read_csv(
    '../dataRepo/ts_ml_mrac_simulation.csv'
).values

# ===== Signals =====
t = data_main[:, 0]

r = data_main[:, 1]
y = data_main[:, 2]
ym = data_main[:, 3]
u = data_main[:, 4]

theta1 = data_main[:, 5]
theta2 = data_main[:, 6]
theta3 = data_main[:, 7]

y0_last = data_main[-1, 8]

r_shifted = r + y0_last
ym_shifted = ym + y0_last

# ===== Panel =====
figPlotParam = fcnDefaultFigSize(14, 0.17, 0.94, 0.06, 0.4, 13)
fig = plt.figure(0, figsize=figPlotParam[0:2])

subPlots = gridspec.GridSpec(3, 1, height_ratios=[2, 1, 1])

ax0 = plt.subplot(subPlots[0])
ax1 = plt.subplot(subPlots[1])
ax2 = plt.subplot(subPlots[2])

# ===== Subplot 1 (y, ym + y0_last, r + y0_last) =====
ax0.plot(t, y, '-', lw=0.6, color='tab:blue', label='y')
ax0.plot(t, ym_shifted, '-', lw=0.6, color='tab:red', label=r'$y_m$')
ax0.plot(t, r_shifted, '--', lw=0.8, color='k', label=r'$r$')

ax0.set_xlabel('čas [s]', ha='left', va='top')
ax0.xaxis.set_label_coords(1.05, -0.07, transform=ax0.transAxes)

ax0.set_ylabel('Výstup [V]', ha='right', va='bottom', rotation=90)
ax0.yaxis.set_label_coords(-0.07, 1.05, transform=ax0.transAxes)

ax0.grid(True, which='both', linestyle='--', linewidth=0.1, color='k')
ax0.legend(frameon=False, fontsize=8, loc='upper right', bbox_to_anchor=(1.13, 1.0))

# ===== Subplot 2 (u) =====
ax1.plot(t, u, '-', lw=0.6, color='tab:orange', label='u')

ax1.set_xlabel('čas [s]', ha='left', va='top')
ax1.xaxis.set_label_coords(1.05, -0.07, transform=ax1.transAxes)

ax1.set_ylabel('Akčný zásah [V]', ha='right', va='bottom', rotation=90)
ax1.yaxis.set_label_coords(-0.07, 1.05, transform=ax1.transAxes)

ax1.grid(True, which='both', linestyle='--', linewidth=0.1, color='k')
ax1.legend(frameon=False, fontsize=8, loc='upper right', bbox_to_anchor=(1.11, 1.0))

# ===== Subplot 3 (theta1, theta2, theta3) =====
ax2.plot(t, theta1, '-', lw=0.8, color='tab:red', label=r'$\theta_1$')
ax2.plot(t, theta2, '-', lw=0.8, color='tab:orange', label=r'$\theta_2$')
ax2.plot(t, theta3, '-', lw=0.8, color='tab:blue', label=r'$\theta_3$')

ax2.set_xlabel('čas [s]', ha='left', va='top')
ax2.xaxis.set_label_coords(1.05, -0.07, transform=ax2.transAxes)

ax2.set_ylabel(r'$\theta$ [-]', ha='right', va='bottom', rotation=90)
ax2.yaxis.set_label_coords(-0.07, 1.05, transform=ax2.transAxes)

ax2.grid(True, which='both', linestyle='--', linewidth=0.1, color='k')
ax2.legend(frameon=False, fontsize=8, loc='upper right', bbox_to_anchor=(1.12 ,1.0))

# ===== Style =====
fcnDefaultLayoutAdj(
    fig,
    figPlotParam[2],
    figPlotParam[3],
    figPlotParam[4],
    figPlotParam[5]
)

plt.savefig(os.path.join(figSaveDir, f"{figName}.png"), dpi=600)
plt.close()
```
