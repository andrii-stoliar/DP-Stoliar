```python
import os
import importlib
import pandas as pd
import myfiglib

importlib.reload(myfiglib)
StackedFigure = myfiglib.StackedFigure

figSaveDir = "../fig"
os.makedirs(figSaveDir, exist_ok=True)
figName = "ts_ml_mrac_real_raw"

data = pd.read_csv("../dataRepo/ts_ml_mrac_real.csv").values

t = data[:, 0]
r, y, ym, u = data[:, 1], data[:, 2], data[:, 3], data[:, 4]
theta1, theta2, theta3 = data[:, 5], data[:, 6], data[:, 7]

y0_last = data[-1, 8]
r_shifted = r + y0_last
ym_shifted = ym + y0_last

fig = StackedFigure(14, 3, [2, 1, 1])

fig.panel(0, "čas [s]", "Výstup [V]")
fig.line(0, t, y, lw=0.1,color="tab:blue", label="y")
fig.line(0, t, ym_shifted, color="tab:red", label=r"$y_m$")
fig.line(0, t, r_shifted, ls="--", lw=0.8, color="k", label=r"$r + y_{pb}$")

fig.panel(1, "čas [s]", "Akčný zásah [V]")
fig.line(1, t, u, color="tab:orange", label="u")

fig.panel(2, "čas [s]", r"$\theta$ [-]")
fig.line(2, t, theta1, lw=0.8, color="tab:red", label=r"$\theta_1$")
fig.line(2, t, theta2, lw=0.8, color="tab:orange", label=r"$\theta_2$")
fig.line(2, t, theta3, lw=0.8, color="tab:blue", label=r"$\theta_3$")

fig.save(os.path.join(figSaveDir, f"{figName}.png"), dpi=600)
```

ALL EXP MRAC comparison

```python
import os
import importlib
import numpy as np
import pandas as pd
import myfiglib

importlib.reload(myfiglib)
StackedFigure = myfiglib.StackedFigure

figSaveDir = "../fig"
os.makedirs(figSaveDir, exist_ok=True)
figName = "ts_mrac_real_comparison"

# -----------------------------
# SETTINGS
# -----------------------------
LW_SIG = 0.18
LW_REF = 0.45

COLOR_ML = "orange"
COLOR_CPP = "blue"
COLOR_PLC = "green"
COLOR_CPP_EXP = "purple"
COLOR_YM = "red"
COLOR_R = "black"


def rmse(a, b):
    return np.sqrt(np.mean((a - b) ** 2))


def load_standard(csv_path, name, color):
    raw = pd.read_csv(csv_path).values

    y0_last = raw[-1, 8]
    data = raw[raw[:, 0] >= 400]

    t = data[:, 0]
    r = data[:, 1]
    y = data[:, 2]
    ym = data[:, 3]
    u = data[:, 4]

    y_shifted = y - y0_last
    u_shifted = u - 3

    return {
        "name": name,
        "color": color,
        "t": t,
        "r": r,
        "ym": ym,
        "y": y_shifted,
        "u": u_shifted
    }


def load_experimental(csv_path, name, color):
    raw = pd.read_csv(csv_path).values

    y0_last = raw[-1, 11]
    data = raw[raw[:, 0] >= 400]

    t = data[:, 0]
    r = data[:, 1]
    y = data[:, 3]
    ym = data[:, 4]
    u = data[:, 7]

    y_shifted = y - y0_last
    u_shifted = u - 3

    return {
        "name": name,
        "color": color,
        "t": t,
        "r": r,
        "ym": ym,
        "y": y_shifted,
        "u": u_shifted
    }


# -----------------------------
# LOAD DATASETS
# -----------------------------
datasets = [
    load_standard("../dataRepo/ts_ml_mrac_real.csv", "MATLAB", COLOR_ML),
    load_standard("../dataRepo/ts_cpp_mrac_real.csv", "Arduino C++", COLOR_CPP),
    load_standard("../dataRepo/ts_plc_mrac_real.csv", "PLC", COLOR_PLC),
    load_experimental("../dataRepo/ts_cpp_mrac_experemental.csv", "C++ ex.", COLOR_CPP_EXP),
]

# -----------------------------
# RMSE PRINT
# -----------------------------
print("RMSE between Δy and Δy_m for each dataset:")
for ds in datasets:
    value = rmse(ds["y"], ds["ym"])
    print(f"{ds['name']}: {value:.6f} V")

# -----------------------------
# FIGURE
# -----------------------------
fig = StackedFigure(9.5, 2, [1, 1])

# ---- panel 1: outputs ----
fig.panel(0, "čas [s]", "Výstup [V]")

fig.line(
    0,
    datasets[0]["t"],
    datasets[0]["r"],
    ls="--",
    lw=LW_REF,
    color=COLOR_R,
    label=r"$\Delta r$"
)

fig.line(
    0,
    datasets[0]["t"],
    datasets[0]["ym"],
    lw=LW_REF,
    color=COLOR_YM,
    label=r"$\Delta y_m$"
)

for ds in datasets:
    fig.line(
        0,
        ds["t"],
        ds["y"],
        lw=LW_SIG,
        color=ds["color"],
        label=rf"$\Delta y$ {ds['name']}"
    )

# ---- panel 2: inputs ----
fig.panel(1, "čas [s]", "Akčný zásah [V]")

for ds in datasets:
    fig.line(
        1,
        ds["t"],
        ds["u"],
        lw=LW_SIG,
        color=ds["color"],
        label=rf"$\Delta u$ {ds['name']}"
    )

fig.save(os.path.join(figSaveDir, f"{figName}.png"), dpi=600)
```
