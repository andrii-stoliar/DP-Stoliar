MRAC real raw experiment

```python
import os
import importlib
import pandas as pd
import myfiglib

importlib.reload(myfiglib)
StackedFigure = myfiglib.StackedFigure

figSaveDir = "../fig"
os.makedirs(figSaveDir, exist_ok=True)
figName = "ts_ml_mrac_real_raw"

data = pd.read_csv("../dataRepo/ts_ml_mrac_real.csv").values

t = data[:, 0]
r, y, ym, u = data[:, 1], data[:, 2], data[:, 3], data[:, 4]
theta1, theta2, theta3 = data[:, 5], data[:, 6], data[:, 7]

y0_last = data[-1, 8]
r_shifted = r + y0_last
ym_shifted = ym + y0_last

fig = StackedFigure(14, 3, [2, 1, 1])

fig.panel(0, "čas [s]", "Výstup [V]")
fig.line(0, t, y, lw=0.1,color="tab:blue", label="y")
fig.line(0, t, ym_shifted, color="tab:red", label=r"$y_m$")
fig.line(0, t, r_shifted, ls="--", lw=0.8, color="k", label=r"$r + y_{pb}$")

fig.panel(1, "čas [s]", "Akčný zásah [V]")
fig.line(1, t, u, color="tab:orange", label="u")

fig.panel(2, "čas [s]", r"$\theta$ [-]")
fig.line(2, t, theta1, lw=0.8, color="tab:red", label=r"$\theta_1$")
fig.line(2, t, theta2, lw=0.8, color="tab:orange", label=r"$\theta_2$")
fig.line(2, t, theta3, lw=0.8, color="tab:blue", label=r"$\theta_3$")

fig.save(os.path.join(figSaveDir, f"{figName}.png"), dpi=600)
```
