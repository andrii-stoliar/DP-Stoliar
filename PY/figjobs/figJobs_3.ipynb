PLC PI real

```python
import os
import importlib
import pandas as pd
import myfiglib
import numpy as np

importlib.reload(myfiglib)
StackedFigure = myfiglib.StackedFigure

figSaveDir = "../fig"
os.makedirs(figSaveDir, exist_ok=True)
figName = "ts_plc_pi_real"

data = pd.read_csv("../dataRepo/ts_plc_pi_real.csv", header=None, skiprows=1).values

t = data[:, 0]
u = data[:, 4] - 4
y = data[:, 1] - data[:, 2]
ref = data[:, 3]

idx_start = np.argmax(t > 600) - 1
idx_end = np.argmax(t > 650)
if idx_end == 0:
    idx_end = len(t)

t0 = t[idx_start]
t = t[idx_start:idx_end] - t0
u = u[idx_start:idx_end]
y = y[idx_start:idx_end]
ref = ref[idx_start:idx_end]

fig = StackedFigure(9.5, 2, [1, 1])

fig.panel(0, "čas [s]", "Výstup [V]", ylim=(-0.1, 0.6))
fig.line(0, t, y, lw=0.6, color="k", label=r"$\Delta y$")
fig.line(0, t, ref, ls="--", lw=0.9, color="red", label="setpoint")

fig.panel(1, "čas [s]", "Vstup [V]", ylim=(-0.5, 2.5))
fig.line(1, t, u, lw=0.6, color="k", label=r"$\Delta u$")

fig.save(os.path.join(figSaveDir, f"{figName}.png"), dpi=300)
```
