```python
import os
import importlib
import numpy as np
import pandas as pd
import myfiglib

importlib.reload(myfiglib)
StackedFigure = myfiglib.StackedFigure

figSaveDir = "../fig"
os.makedirs(figSaveDir, exist_ok=True)

figName = "ts_ml_prep_simulation"

data_main = pd.read_csv("../dataRepo/ts_ml_prep_simulation.csv").values

t = data_main[:, 0]

r_l = data_main[:, 1]
r_g = data_main[:, 2]
r_gf = data_main[:, 3]

y = data_main[:, 4]
y_pb = data_main[:, 5]

u = data_main[:, 6]
u_g = data_main[:, 7]
u_l = data_main[:, 8]
u_pb = data_main[:, 9]

u_g_clipped = np.maximum(u_g, 0)
u_l_clipped = np.maximum(u_l, 0)
u_pb_clipped = np.maximum(u_pb, 0)

active_g = data_main[:, 10]

Ki_l = data_main[:, 11]
Kp_l = data_main[:, 12]

r_sum = r_gf + r_l

fig = StackedFigure(
    height_cm=22,
    nrows=6,
    height_ratios=[2, 1, 1, 1, 1, 1]
)

fig.panel(0, xlabel="čas [s]", ylabel="Výstup [V]")
fig.line(0, t, y, ls="-", lw=0.6, color="tab:blue", label="y")
fig.line(0, t, y_pb, ls="-", lw=0.6, color="tab:red", label=r"$y_{pb}$")
fig.line(0, t, r_sum, ls="--", lw=0.8, color="k", label=r"$r_{gf} + r_l$")

fig.panel(1, xlabel="čas [s]", ylabel="Referencie [V]")
fig.line(1, t, r_l, ls="-", lw=0.6, color="tab:blue", label=r"$r_l$")
fig.line(1, t, r_g, ls="-", lw=0.6, color="tab:orange", label=r"$r_g$")
fig.line(1, t, r_gf, ls="-", lw=0.6, color="tab:red", label=r"$r_{gf}$")

fig.panel(2, xlabel="čas [s]", ylabel="Akčný zásah [V]")
fig.line(2, t, u, ls="-", lw=0.6, color="tab:orange", label="u")

fig.panel(3, xlabel="čas [s]", ylabel="Zložky riadenia [V]")
fig.line(3, t, u_g_clipped, ls="-", lw=0.6, color="tab:red", label=r"$u_g$")
fig.line(3, t, u_l_clipped, ls="-", lw=0.6, color="tab:blue", label=r"$u_l$")
fig.line(3, t, u_pb_clipped, ls="-", lw=0.6, color="tab:green", label=r"$u_{pb}$")

fig.panel(4, xlabel="čas [s]", ylabel="Parametre [-]")
fig.line(4, t, Kp_l, ls="-", lw=0.8, color="tab:blue", label=r"$K_{p,l}$")
fig.line(4, t, Ki_l, ls="-", lw=0.8, color="tab:red", label=r"$K_{i,l}$")

fig.panel(5, xlabel="čas [s]", ylabel="Aktivácia [-]")
fig.line(5, t, active_g, ls="-", lw=0.8, color="tab:green", label=r"$active_g$")

fig.save(os.path.join(figSaveDir, f"{figName}.png"), dpi=600)
```

```python
import os
import importlib
import pandas as pd
import myfiglib

importlib.reload(myfiglib)
StackedFigure = myfiglib.StackedFigure

figSaveDir = "../fig"
os.makedirs(figSaveDir, exist_ok=True)
figName = "ts_ml_prep_real_zoom"

data = pd.read_csv("../dataRepo/ts_ml_prep_real.csv").values

t = data[:, 0]
r_l, r_g = data[:, 1], data[:, 2]
y, y_pb = data[:, 4], data[:, 5]

mask1 = (t >= 180) & (t <= 399.9)
mask2 = (t >= 580) & (t <= 799.9)

t1, t2 = t[mask1], t[mask2]
y1, y2 = y[mask1], y[mask2]
y_pb1, y_pb2 = y_pb[mask1], y_pb[mask2]
r_l1, r_l2 = r_l[mask1], r_l[mask2]
r_g1, r_g2 = r_g[mask1], r_g[mask2]

rg_upper1, rg_lower1 = 1.03 * r_g1, 0.97 * r_g1
rg_upper2, rg_lower2 = 1.03 * r_g2, 0.97 * r_g2

y_ref1 = y_pb1 + r_l1
y_ref2 = y_pb2 + r_l2

fig = StackedFigure(12, 2, [1, 1])

fig.panel(0, "čas [s]", "Výstup [V]")
fig.axes[0].fill_between(t1, rg_lower1, rg_upper1, color="tab:green", alpha=0.15)
fig.line(0, t1, y1, color="tab:blue", label="y")
fig.line(0, t1, y_pb1, color="tab:red", label=r"$y_{pb}$")
fig.line(0, t1, y_ref1, ls="--", lw=0.8, color="k", label=r"$y_{pb} + r_l$")
fig.line(0, t1, r_g1, ls=":", lw=0.8, color="tab:green", label=r"$r_g$")

fig.panel(1, "čas [s]", "Výstup [V]")
fig.axes[1].fill_between(t2, rg_lower2, rg_upper2, color="tab:green", alpha=0.15)
fig.line(1, t2, y2, color="tab:blue", label="y")
fig.line(1, t2, y_pb2, color="tab:red", label=r"$y_{pb}$")
fig.line(1, t2, y_ref2, ls="--", lw=0.8, color="k", label=r"$y_{pb} + r_l$")
fig.line(1, t2, r_g2, ls=":", lw=0.8, color="tab:green", label=r"$r_g$")

fig.save(os.path.join(figSaveDir, f"{figName}.png"), dpi=600)
```
