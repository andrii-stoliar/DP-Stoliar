ALL EXP PREP comparison 

```python
import os
import importlib
import numpy as np
import pandas as pd
import myfiglib

importlib.reload(myfiglib)
StackedFigure = myfiglib.StackedFigure

figSaveDir = "../fig"
os.makedirs(figSaveDir, exist_ok=True)
figName = "ts_prep_real_comparison"

# -----------------------------
# SETTINGS
# -----------------------------
LW_SIG = 0.45
LW_REF = 0.9

COLOR_MATLAB = "orange"
COLOR_CPP = "blue"
COLOR_PLC = "green"
COLOR_REF = "red"


def rmse(y, ref):
    return np.sqrt(np.mean((y - ref) ** 2))


def load_dataset(csv_path, name, color):
    data = pd.read_csv(csv_path).values

    t = data[:, 0]
    r_l = data[:, 1]
    r_g = data[:, 2]
    r_gf = data[:, 3]
    y = data[:, 4]
    y_pb = data[:, 5]
    u = data[:, 6]
    u_g = data[:, 7]
    u_l = data[:, 8]
    u_pb = data[:, 9]
    active_g = data[:, 10]
    Ki_l = data[:, 11]
    Kp_l = data[:, 12]

    # if needed in future
    u_g = np.maximum(u_g, 0)
    u_l = np.maximum(u_l, 0)
    u_pb = np.maximum(u_pb, 0)

    r_sum = r_gf + r_l

    return {
        "name": name,
        "color": color,
        "t": t,
        "y": y,
        "u": u,
        "r_sum": r_sum,
        "y_pb": y_pb,
        "r_l": r_l,
        "r_g": r_g,
        "r_gf": r_gf,
        "u_g": u_g,
        "u_l": u_l,
        "u_pb": u_pb,
        "active_g": active_g,
        "Ki_l": Ki_l,
        "Kp_l": Kp_l
    }


datasets = [
    load_dataset("../dataRepo/ts_ml_prep_real.csv", "MATLAB", COLOR_MATLAB),
    load_dataset("../dataRepo/ts_cpp_prep_real.csv", "Arduino C++", COLOR_CPP),
    load_dataset("../dataRepo/ts_plc_prep_real.csv", "PLC", COLOR_PLC),
]

# -----------------------------
# RMSE print
# -----------------------------
print("RMSE relative to each dataset's own (r_gf + r_l):")
for ds in datasets:
    value = rmse(ds["y"], ds["r_sum"])
    print(f"{ds['name']}: {value:.6f} V")

# -----------------------------
# FIGURE
# -----------------------------
fig = StackedFigure(10, 2, [1, 1])

# ---- panel 1: y + one common r_sum ----
fig.panel(0, "čas [s]", "Výstup [V]")

# common reference only for visualization
fig.line(
    0,
    datasets[0]["t"],
    datasets[0]["r_sum"],
    ls="--",
    lw=LW_REF,
    color=COLOR_REF,
    label=r"$r_{gf} + r_l$"
)

for ds in datasets:
    fig.line(
        0,
        ds["t"],
        ds["y"],
        lw=LW_SIG,
        color=ds["color"],
        label=rf"$y$ {ds['name']}"
    )

# ---- panel 2: u from all experiments ----
fig.panel(1, "čas [s]", "Akčný zásah [V]")

for ds in datasets:
    fig.line(
        1,
        ds["t"],
        ds["u"],
        lw=LW_SIG,
        color=ds["color"],
        label=rf"$u$ {ds['name']}"
    )

fig.save(os.path.join(figSaveDir, f"{figName}.png"), dpi=600)
```
