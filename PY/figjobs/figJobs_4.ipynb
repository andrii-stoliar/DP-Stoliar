```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from figFcns_TeX import *

# ===== Params =====
figSaveDir = '../fig'
os.makedirs(figSaveDir, exist_ok=True)

figName = 'ts_ml_prep_real_zoom'

# ===== Load CSV =====
data_main = pd.read_csv(
    '../dataRepo/ts_ml_prep_real.csv'
).values

# ===== Signals =====
t = data_main[:, 0]

r_l = data_main[:, 1]
r_g = data_main[:, 2]

y = data_main[:, 4]
y_pb = data_main[:, 5]

# ===== Windows =====
mask1 = (t >= 180) & (t <= 399.9)
mask2 = (t >= 580) & (t <= 799.9)

t1 = t[mask1]
t2 = t[mask2]

y1 = y[mask1]
y2 = y[mask2]

y_pb1 = y_pb[mask1]
y_pb2 = y_pb[mask2]

r_l1 = r_l[mask1]
r_l2 = r_l[mask2]

r_g1 = r_g[mask1]
r_g2 = r_g[mask2]

rg_upper1 = r_g1 * 1.03
rg_lower1 = r_g1 * 0.97

rg_upper2 = r_g2 * 1.03
rg_lower2 = r_g2 * 0.97

y_pb_plus_rl_1 = y_pb1 + r_l1
y_pb_plus_rl_2 = y_pb2 + r_l2

# ===== Panel =====
figPlotParam = fcnDefaultFigSize(12, 0.17, 0.93, 0.07, 0.4, 13)
fig = plt.figure(0, figsize=figPlotParam[0:2])

# ===== Subplot height ratios =====
subplotHeightRatios = [1, 1]

subPlots = gridspec.GridSpec(2, 1, height_ratios=subplotHeightRatios)

ax0 = plt.subplot(subPlots[0])
ax1 = plt.subplot(subPlots[1])

# ===== Subplot 1 (180–400 s) =====
# tolerance band
ax0.fill_between(
    t1,
    rg_lower1,
    rg_upper1,
    color='tab:green',
    alpha=0.15
)

ax0.plot(t1, y1, '-', lw=0.6, color='tab:blue', label='y')
ax0.plot(t1, y_pb1, '-', lw=0.6, color='tab:red', label=r'$y_{pb}$')
ax0.plot(t1, y_pb_plus_rl_1, '--', lw=0.8, color='k', label=r'$y_{pb} + r_l$')
ax0.plot(t1, r_g1, ':', lw=0.8, color='tab:green', label=r'$r_g$')

ax0.set_xlabel('čas [s]', ha='left', va='top')
ax0.xaxis.set_label_coords(1.05, -0.07, transform=ax0.transAxes)

ax0.set_ylabel('Výstup [V]', ha='right', va='bottom', rotation=90)
ax0.yaxis.set_label_coords(-0.07, 1.05, transform=ax0.transAxes)

ax0.grid(True, which='both', linestyle='--', linewidth=0.1, color='k')
ax0.legend(frameon=False, fontsize=8, loc='upper right', bbox_to_anchor=(1.16, 1.0))

# ===== Subplot 2 (580–800 s) =====
# tolerance band
ax1.fill_between(
    t2,
    rg_lower2,
    rg_upper2,
    color='tab:green',
    alpha=0.15
)

ax1.plot(t2, y2, '-', lw=0.6, color='tab:blue', label='y')
ax1.plot(t2, y_pb2, '-', lw=0.6, color='tab:red', label=r'$y_{pb}$')
ax1.plot(t2, y_pb_plus_rl_2, '--', lw=0.8, color='k', label=r'$y_{pb} + r_l$')
ax1.plot(t2, r_g2, ':', lw=0.8, color='tab:green', label=r'$r_g$')

ax1.set_xlabel('čas [s]', ha='left', va='top')
ax1.xaxis.set_label_coords(1.05, -0.07, transform=ax1.transAxes)

ax1.set_ylabel('Výstup [V]', ha='right', va='bottom', rotation=90)
ax1.yaxis.set_label_coords(-0.07, 1.05, transform=ax1.transAxes)

ax1.grid(True, which='both', linestyle='--', linewidth=0.1, color='k')
ax1.legend(frameon=False, fontsize=8, loc='upper right', bbox_to_anchor=(1.16, 1.0))

# ===== Style =====
fcnDefaultLayoutAdj(
    fig,
    figPlotParam[2],
    figPlotParam[3],
    figPlotParam[4],
    figPlotParam[5]
)

plt.savefig(os.path.join(figSaveDir, f"{figName}.png"), dpi=600)
plt.close()
```
